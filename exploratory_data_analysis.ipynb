```python
import os 
import cv2
from PIL import Image
import matplotlib.pyplot as plt
```

```python
#
# Check the sizes of the processed images
# Image size is 256x256
#

training = "cleaned/Training"
testing = "cleaned/Testing"
training_dir = os.listdir(training)
testing_dir = os.listdir(testing)
IMG_SIZE = 256
# Initialize dictionaries to store folder sizes
training_sizes = {}
testing_sizes = {}

for dir in training_dir:
    save_path = 'cleaned/Training/'+ dir
    path = os.path.join(training,dir)
    image_dir = os.listdir(path)
    training_sizes[dir] = len(image_dir)
    print(f"Folder name: {path}, Size: {len(image_dir)}")
    for img in image_dir:
        #image = cv2.imread(os.path.join(path,img))
        image = Image.open(os.path.join(path,img))
        print(f"Filename: {path}\\{img}, Size: {image.size}, Mode: {image.mode}, Format: {image.format}")

for dir in testing_dir:
    save_path = 'cleaned/Testing/'+ dir
    path = os.path.join(testing,dir)
    image_dir = os.listdir(path)
    testing_sizes[dir] = len(image_dir)
    print(f"Folder name: {path}, Size: {len(image_dir)}")
    for img in image_dir:
        image = Image.open(os.path.join(path,img))
        print(f"Filename: {path}\\{img}, Size: {image.size}, Mode: {image.mode}, Format: {image.format}")



# Plot pie chart for training data
plt.figure(figsize=(8, 6))
plt.pie(training_sizes.values(), labels=training_sizes.keys(), autopct='%1.1f%%', startangle=140)
plt.title('Training Data Distribution')
plt.show()

# Plot pie chart for testing data
plt.figure(figsize=(8, 6))
plt.pie(testing_sizes.values(), labels=testing_sizes.keys(), autopct='%1.1f%%', startangle=140)
plt.title('Testing Data Distribution')
plt.show()
```
